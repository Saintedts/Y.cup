```python
import pandas as pd
```

```python
desired_percentage = 50
```

```python
df = pd.read_csv('feature_files/feature_1.csv',names=["ID",1])

for i in range(2,21):
    temp = pd.read_csv(f'feature_files/feature_{i}.csv', names=["ID",i ])
    df = df.merge(temp,how='outer', left_on='ID', right_on='ID')

df
```

```python
qty_users = 0

for i in range(0,len(df)):
    count = 0
    for j in range(1,len(df.columns)):
        if pd.notnull(df.iloc[i,j]):
            count = count + 1
    if (count / (len(df.columns) - 1) * 100) >=  desired_percentage:
        qty_users = qty_users + 1

qty_users
```
